```python
def execute_sql(sql, params=None):
    """
    For creating tables, inserting into tables.
    """
    with psycopg.connect(conninfo) as conn:
        with conn.cursor() as cur:
            cur.execute(sql, params)
        conn.commit()
```

```python
def query_df(sql, params=None):
    """
    For running queries. Returns pandas dataframe of query ouput.
    """
    with psycopg.connect(conninfo) as conn:
        with conn.cursor() as cur:
            cur.execute(sql, params)
            cols = [d.name for d in cur.description]
            return pd.DataFrame(cur.fetchall(), columns=cols)
```

```python
query = """
WITH calc_overall_avg_rank AS(
SELECT nd2324.player_name, 
    (nd2324.player_rank_per_season + nd2425.player_rank_per_season) / 2 AS avg_player_rank_over_two_seasons
FROM nba_data_2023_2024 nd2324
LEFT JOIN nba_data_2024_2025 nd2425
ON nd2324.player_name = nd2425.player_name
WHERE nd2324.player_rank_per_season IS NOT NULL and nd2425.player_rank_per_season IS NOT NULL
GROUP BY nd2324.player_name, nd2324.player_rank_per_season, nd2425.player_rank_per_season
)
SELECT coar.player_name, coar.avg_player_rank_over_two_seasons
FROM calc_overall_avg_rank coar
ORDER BY coar.avg_player_rank_over_two_seasons ASC
LIMIT 5;
"""
```

2. top 5 players by different shooting attempts that has shooting attempts greater than the mean shooting attempt for each shot

```python
shooting_attempts = ['field_goals_attempts', 'two_pointer_attempts', 'three_pointer_attempts', 'field_goals_attempts', 'free_throw_attempts']
shooting_percentage = ['field_goals_percentage', 'two_pointer_percentage', 'three_pointer_percentage', 'effective_field_goal_percentage', 'free_throw_percentage']
attempt_naming = ['avg_fga_over_two_seasons', 'avg_tpa_over_two_seasons', 'avg_tpa_over_two_seasons', 'avg_fga_over_two_seasons', 'avg_fta_over_two_seasons']
percentage_naming = ['avg_fgp_over_two_seasons', 'avg_tpp_over_two_seasons', 'avg_tpp_over_two_seasons', 'avg_efgp_over_two_seasons', 'avg_ftp_over_two_seasons']
```

```python
for i in range(len(shooting_attempts)):
    query = """
    WITH calc_avg_fga AS (
    SELECT AVG( """ + shooting_attempts[i] + """) AS average_fga
    FROM nba_data
    WHERE """ + shooting_attempts[i] + """ IS NOT NULL AND """ + shooting_attempts[i] + """ > 0
    ),
    calc_overall_field_goals_attempts AS(
    SELECT nd2324.player_name, nd2324.team,
    (nd2324.""" + shooting_attempts[i] + """ + nd2425.""" + shooting_attempts[i] + """) / 2 AS """ + attempt_naming[i] + """,
    (nd2324.""" + shooting_percentage[i] + """ + nd2425.""" + shooting_percentage[i] + """) / 2 AS """ + percentage_naming[i] + """
    FROM nba_data_2023_2024 nd2324
    LEFT JOIN nba_data_2024_2025 nd2425
    ON nd2324.player_name = nd2425.player_name
    WHERE nd2324.player_rank_per_season IS NOT NULL and nd2425.player_rank_per_season IS NOT NULL
    GROUP BY nd2324.player_name, nd2324.team, nd2324.""" + shooting_attempts[i] + """, nd2425.""" + shooting_attempts[i] + """, nd2324.""" + shooting_percentage[i] + """, nd2425.""" + shooting_percentage[i] + """
    )
    SELECT cofga.player_name, cofga.team,
    ROUND(cofga.""" + percentage_naming[i] + """ * 100, 2) AS """ + percentage_naming[i] + """
    FROM calc_overall_field_goals_attempts cofga
    CROSS JOIN calc_avg_fga caf
    WHERE cofga.""" + attempt_naming[i] + """ > caf.average_fga
    ORDER BY cofga.""" + percentage_naming[i] + """ DESC
    LIMIT 5;
    """

    df = query_df(query)
    display(df)
```

##### Team Momentum calculation
Aggregate team stats across two season, and compute the momentum in statistics (based on recruiter's needs)

```python
recruters_needs_shooting = ['field_goals', 'three_pointer', 'two_pointer', 'effective_field_goal', 'free_throw']
recruters_needs_non_shooting = ['mins_played', 'field_goals', 'offensive_rebound', 'defensive_rebound', 'total_rebound', 'assist', 'steal', 'turnover']
```

```python
# shooting stats query
shooting_query = ""
for s in recruters_needs_shooting:
    if s == 'effective_field_goal':
        query_partial = "ROUND(AVG(effective_field_goal_percentage) * 100 , 2) AS avg_effective_field_goal_percent,"
        shooting_query += query_partial
    else:
        query_partial = "ROUND((SUM(" + s + ")*1.0 / SUM(" + s + "_attempts)) * 100, 2) AS avg_" + s + "_percent,"
        shooting_query += query_partial

# non shooting stats query
non_shooting_query = ""
for ns in recruters_needs_non_shooting:
    if ns == 'field_goals':
        query_partial = "SUM(field_goals) AS total_field_goals,"
        non_shooting_query += query_partial
    else:
        query_partial = "ROUND(AVG(" + ns + ") , 2) AS avg_" + ns + ","
        non_shooting_query += query_partial
non_shooting_query = non_shooting_query[:-1]

# combine
seasonal_selected_stat = shooting_query + non_shooting_query
seasonal_selected_stat

########################################################################################################################################

# final momentum select part
# shooting stats query
final_select_part = ""
for s in recruters_needs_shooting:
    if s == 'field_goals':
        query_partial = "nd2425.total_field_goals - nd2324.total_field_goals AS field_goal_momentum,"
        final_select_part += query_partial
    else:
        query_partial = "nd2425.avg_" + s + "_percent - nd2324.avg_" + s + "_percent AS " + s + "_perct_momentum,"
        final_select_part += query_partial

# nonshooting stats query
for ns in recruters_needs_non_shooting:
    if ns == 'field_goals':
        query_partial = "nd2425.total_field_goals - nd2324.total_field_goals AS field_goal_momentum,"
        non_shooting_query += query_partial
    else:
        query_partial = "nd2425.avg_" + ns + " - nd2324.avg_" + ns + " AS " + ns + "_momentum,"
        final_select_part += query_partial
final_select_part = final_select_part[:-1]
final_select_part
```

```python
print(seasonal_selected_stat)
```

```python
query = """
WITH nd2324 AS (
SELECT team, 
""" + seasonal_selected_stat + """
FROM nba_data_2023_2024
GROUP BY team
),
nd2425 AS(
SELECT team,
""" + seasonal_selected_stat + """
FROM nba_data_2024_2025
GROUP BY team
)
SELECT nd2324.team,
""" + final_select_part + """
FROM nd2324
LEFT JOIN nd2425
ON nd2324.team = nd2425.team
ORDER BY field_goal_momentum DESC, assist_momentum DESC
"""
```

for instance, if a recuiter is trying to find a player with scoring ability, assist ability, and plays aggresive on the defensive side (high personal foul), we will only be looking at 
stats such as:

 points, field_goals_percentage, three_pointer_percentage, two_pointer_percentage, effective_field_goal_percentage, assist, and personal_fouls.

```python
# meed to redefine players that matches the recruiter's goals
new_candidates = []
# these were the top shooters but removed top free throw shooter
new_candidates.append("Luke Kennard")
new_candidates.append("Brandon Clarke")
new_candidates.append("Brandon Clarke")
# these were players with top assist and personal_fouls.
new_candidates.append("Ja Morant")
new_candidates.append("Jaren Jackson Jr.")


# where piece in the sql query to filter out the players we want to see
player_filtering_query = """player_name = """ 
for i in range(len(new_candidates)):
    if i < len(new_candidates) - 1:
        player_filtering_query += """'""" + new_candidates[i] + """' OR  player_name = """
    else:
        player_filtering_query += """'""" + new_candidates[i] + """' AND team != '2TM' AND team != '3TM'"""
    
player_filtering_query
```

```python
stats_needed = ['points', 'field_goals_percentage', 'three_pointer_percentage', 'two_pointer_percentage', 'effective_field_goal_percentage', 'assist', 'personal_foul']
```

```python
select_query = ""

for s in stats_needed:
    if s == "games" or s == "games_started":
        select_query +=  "SUM(" + s + ") AS " + s + ","
    elif "percentage" in s and s != "effective_field_goal_percentage":
        select_query += "ROUND(AVG(" + s + ") * 100, 2) AS " + s + ","
    elif s == "effective_field_goal_percentage":
        select_query += "ROUND(AVG(effective_field_goal_percentage) * 100, 2) AS effective_field_goal_percentage,"
    else:
        select_query +=  "ROUND(AVG(" + s + ") / SUM(games), 2) AS " + s + ","

select_query = select_query[:-1]
select_query
```

```python
final_select_query = ""

for s in stats_needed:
    final_select_query += "COALESCE(nd2425." + s + ", 0) - COALESCE(nd2324." + s + ", 0) AS " + s + ","

final_select_query = final_select_query[:-1]
final_select_query
```

```python
query = """
WITH nd2324_player_stats AS (
SELECT player_name,
    STRING_AGG(DISTINCT team, ', ') AS teams_2324,
    """ + select_query + """
FROM nba_data_2023_2024
WHERE """ + player_filtering_query + """
GROUP BY player_name
),
nd2425_player_stats AS (
SELECT player_name, 
    STRING_AGG(DISTINCT team, ', ') AS teams_2425,
    """ + select_query + """
FROM nba_data_2024_2025
WHERE """ + player_filtering_query + """
GROUP BY player_name
) 
SELECT COALESCE(nd2324.player_name, nd2425.player_name) AS player,
    nd2324.teams_2324 AS s2324team,
    nd2425.teams_2425 AS s2425team,
    """ + final_select_query + """
FROM nd2324_player_stats nd2324
INNER JOIN nd2425_player_stats nd2425
ON nd2324.player_name = nd2425.player_name
"""
```

#### Candidate Scoring and ranking

```python
stats_needed = ['points', 'field_goals_percentage', 'three_pointer_percentage', 'two_pointer_percentage', 'effective_field_goal_percentage', 'assist', 'personal_foul']
player_stat_query = ""
for sn in stats_needed:
    player_stat_query += sn + ","

player_stat_query = player_stat_query[:-1]
player_stat_query
```

```python
ranking_query = ""
for sn in stats_needed:
    ranking_query += "RANK() OVER (ORDER BY (" + sn + ") ASC) AS " + sn + "_rk,"

ranking_query = ranking_query[:-1]
ranking_query
```

```python
final_select_ind_rank = ""
final_select_add_point = ""
for sn in stats_needed:
    final_select_ind_rank += sn + "_rk, "
    final_select_add_point += sn + "_rk + "

final_select_ind_rank = final_select_ind_rank
final_select_add_point = final_select_add_point[:-3]
final_select_add_point += " AS total_score"

# combine
rank_sel_query = final_select_ind_rank + final_select_add_point
rank_sel_query
```

```python
query = """
WITH player_stats AS (
SELECT player_rank_per_season AS rank,
    player_name, pos,
    """ + player_stat_query + """
FROM nba_data_2024_2025
WHERE """ + player_filtering_query + """
),
each_stat_rank AS (
SELECT rank, player_name, pos,
    """ + ranking_query + """
FROM player_stats
)
SELECT rank, player_name, pos,
    """ + rank_sel_query + """
FROM each_stat_rank
ORDER BY total_score DESC
"""
```

```python
df = query_df(query)
```
