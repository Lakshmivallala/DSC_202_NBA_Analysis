```python
def execute_sql(sql, params=None):
    """
    For creating tables, inserting into tables.
    """
    with psycopg.connect(conninfo) as conn:
        with conn.cursor() as cur:
            cur.execute(sql, params)
        conn.commit()
```

```python
def query_df(sql, params=None):
    """
    For running queries. Returns pandas dataframe of query ouput.
    """
    with psycopg.connect(conninfo) as conn:
        with conn.cursor() as cur:
            cur.execute(sql, params)
            cols = [d.name for d in cur.description]
            return pd.DataFrame(cur.fetchall(), columns=cols)
```

```python
query = """
WITH calc_overall_avg_rank AS(
SELECT nd2324.player_name, 
    (nd2324.player_rank_per_season + nd2425.player_rank_per_season) / 2 AS avg_player_rank_over_two_seasons
FROM nba_data_2023_2024 nd2324
LEFT JOIN nba_data_2024_2025 nd2425
ON nd2324.player_name = nd2425.player_name
WHERE nd2324.player_rank_per_season IS NOT NULL and nd2425.player_rank_per_season IS NOT NULL
GROUP BY nd2324.player_name, nd2324.player_rank_per_season, nd2425.player_rank_per_season
)
SELECT coar.player_name, coar.avg_player_rank_over_two_seasons
FROM calc_overall_avg_rank coar
ORDER BY coar.avg_player_rank_over_two_seasons ASC
LIMIT 5;
"""
```

##### Compare seasonal stats (23-24 season vs 24-25 season)

1. Find top scorer on each team
    * 2023-2024 season
    * 2024-2025 season

```python
query = """
WITH season_23_24_highest_score AS (
WITH calc_point_ranks AS(
SELECT team, player_name, points, 
    RANK() OVER (PARTITION BY team ORDER BY points DESC) AS points_rank
FROM nba_data_2023_2024
GROUP BY team, player_name, points
ORDER BY team, points DESC
)
SELECT team, player_name, points
FROM calc_point_ranks
WHERE points_rank = 1
ORDER BY team
),
season_24_25_highest_score AS (
WITH calc_point_ranks AS(
SELECT team, player_name, points, 
    RANK() OVER (PARTITION BY team ORDER BY points DESC) AS points_rank
FROM nba_data_2024_2025
GROUP BY team, player_name, points
ORDER BY team, points DESC
)
SELECT team, player_name, points
FROM calc_point_ranks
WHERE points_rank = 1
ORDER BY team
)
SELECT COALESCE(s1hs.team, s2hs.team) AS team,
    s1hs.player_name, s1hs.points AS s_23_24_points, s2hs.player_name, s2hs.points AS s_24_25_points
FROM season_23_24_highest_score s1hs
FULL OUTER JOIN season_24_25_highest_score s2hs
ON s1hs.team = s2hs.team AND s1hs.player_name = s2hs.player_name
WHERE COALESCE(s1hs.team, s2hs.team) = 'MEM'
ORDER BY s1hs.team, s1hs.player_name, s2hs.player_name 
"""
```

```python
df = query_df(query)
```

### closing case find the best matched player

```python
query = """
WITH player_stats AS (
SELECT player_rank_per_season AS rank,
    player_name, pos,
    ROUND(points / games, 2) AS pts_per_game,
    field_goals, field_goals_attempts AS fg_atmpt, field_goals_percentage AS fg_perct,
    three_pointer, three_pointer_attempts AS three_pt_atmpt, three_pointer_percentage AS three_pt_perct, 
    two_pointer, two_pointer_attempts AS two_pt_atmpt, two_pointer_percentage AS two_pt_perct, 
    assist, personal_foul
FROM nba_data_2024_2025
WHERE player_name = 'Jaren Jackson Jr.' OR  player_name = 'Jay Huff' OR player_name = 'Luke Kennard' OR player_name = 'Brandon Clarke' OR player_name = 'Ja Morant'
),
each_stat_rank AS (
SELECT rank, player_name, pos,
    RANK() OVER (ORDER BY (pts_per_game) ASC) AS pts_per_game_rk,
    RANK() OVER (ORDER BY (fg_perct) ASC) AS fg_perct_rk,
    RANK() OVER (ORDER BY (three_pt_perct) ASC) AS three_pt_perct_rk,
    RANK() OVER (ORDER BY (two_pt_perct) ASC) AS two_pt_perct_rk,
    RANK() OVER (ORDER BY (assist) ASC) AS assist_rk,
    RANK() OVER (ORDER BY (personal_foul) ASC) AS personal_foul_rk
FROM player_stats
)
SELECT rank, player_name, pos, pts_per_game_rk, fg_perct_rk, three_pt_perct_rk, two_pt_perct_rk, assist_rk, personal_foul_rk,
    pts_per_game_rk + fg_perct_rk + three_pt_perct_rk + two_pt_perct_rk + assist_rk + personal_foul_rk AS total_score
FROM each_stat_rank
ORDER BY total_score DESC
"""
```
